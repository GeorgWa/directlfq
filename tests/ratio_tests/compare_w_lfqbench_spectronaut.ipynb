```python
#input_file = "../../test_data/system_tests/lfqbench/spectronaut/20220601_115613_TTOF6600_Report.tsv.spectronaut_fragion_isotopes.aq_reformat.tsv"
input_file = "../../test_data/system_tests/lfqbench/spectronaut/20220601_115613_TTOF6600_Report.tsv"
maxquant_file = "../../test_data/system_tests/lfqbench/maxDIA/proteinGroups.txt"
spectronaut_protein_file = "../../test_data/system_tests/lfqbench/spectronaut/20220601_115613_TTOF6600_Report.tsv.spectronaut_protein.aq_reformat.tsv"

#directlfq_output = "../../test_data/system_tests/lfqbench/spectronaut/20220601_115613_TTOF6600_Report.tsv.spectronaut_fragion_isotopes.aq_reformat.tsv.protein_intensities.tsv"

directlfq_output = "../../test_data/system_tests/lfqbench/spectronaut/20220601_115613_TTOF6600_Report.tsv.protein_intensities.tsv"

samplemap_spectronaut ="../../test_data/system_tests/lfqbench/spectronaut/samplemap.tsv"
samplemap_directlfq = "../../test_data/system_tests/lfqbench/spectronaut/samplemap.tsv"
samplemap_maxquant = "../../test_data/system_tests/lfqbench/maxDIA/samplemap.maxLFQ.tsv"


run_directlfq = True
```

```python
if run_directlfq:
    import directlfq.lfq_manager as lfqmgr
    lfqmgr.run_lfq(input_file=input_file, input_type_to_use="spectronaut_precursor")

```

```python
import directlfq.benchmarking as lfqbenchmark
import directlfq.utils as lfqutils
import pandas as pd

samples_used_directlfq = lfqutils.get_samples_used_from_samplemap_file(samplemap_directlfq, cond1="A", cond2="B")
samples_used_spectronaut = lfqutils.get_samples_used_from_samplemap_file(samplemap_spectronaut, cond1="A", cond2="B")
samples_used_maxlfq = lfqutils.get_samples_used_from_samplemap_file(samplemap_maxquant, cond1="A", cond2 = "B")


restable_directlfq = lfqbenchmark.ResultsTableDirectLFQ(input_file=directlfq_output, input_name="directLFQ", samples_c1=samples_used_directlfq[0], samples_c2=samples_used_directlfq[1])
restable_spectronaut = lfqbenchmark.ResultsTableDirectLFQ(input_file=spectronaut_protein_file, input_name="spectronaut", samples_c1=samples_used_directlfq[0], samples_c2=samples_used_directlfq[1])
restable_maxlfq = lfqbenchmark.ResultsTableMaxQuant(input_file=maxquant_file, input_name="maxLFQ", samples_c1=samples_used_maxlfq[0], samples_c2=samples_used_maxlfq[1])
```

```python
display(restable_directlfq.formated_dataframe)
display(restable_spectronaut.formated_dataframe)
display(restable_maxlfq.formated_dataframe)
```

```python
organism_annotator_mq = lfqbenchmark.OrganismAnnotatorMaxQuant(mapping_file=maxquant_file)
organism_annotator_mq.annotate_table_with_organism(restable_maxlfq)
display(restable_maxlfq.formated_dataframe)
```

```python
organism_annotator_directlfq = lfqbenchmark.OrganismAnnotatorSpectronaut(mapping_file = spectronaut_protein_file, protein_column="protein", organism_column="ion")
organism_annotator_directlfq.annotate_table_with_organism(restable_spectronaut)
organism_annotator_directlfq.annotate_table_with_organism(restable_directlfq)
```
