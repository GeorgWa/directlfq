```python
mq_evidence = "../../test_data/system_tests/organellar_maps/evidence.txt"
mq_protein_groups = "../../test_data/system_tests/organellar_maps/proteinGroups.txt"
```

```python
import pandas as pd

evidence_df = pd.read_csv(mq_evidence, sep="\t")
protein_groups_df = pd.read_csv(mq_protein_groups, sep="\t")
```

```python
display(evidence_df)
```

```python
def annotate_protein_id_to_evidence_txt(evidence_df, protein_groups_df):
    evidence_id2protein_map = create_evidence_id2protein_map(protein_groups_df)
    evidence_df["Protein IDs"] = evidence_df["id"].astype("str").map(evidence_id2protein_map)
    return evidence_df

def create_evidence_id2protein_map(protein_groups_df):
    evidence_id2protein_map = {}
    for protein_id, evidence_ids in zip(protein_groups_df["Protein IDs"], protein_groups_df["Evidence IDs"]):
        evidence_ids = evidence_ids.split(";")
        for evidence_id in evidence_ids:
            evidence_id2protein_map[evidence_id] = protein_id
    return evidence_id2protein_map


evidence_df_annotated = annotate_protein_id_to_evidence_txt(evidence_df, protein_groups_df)

display(evidence_df_annotated)


evidence_df_annotated.to_csv(mq_evidence.replace("evidence.txt", "evidence_annotated.txt"), sep="\t", index=False)
```
