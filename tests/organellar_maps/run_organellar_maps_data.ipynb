```python
mq_evidence = "../../test_data/system_tests/organellar_maps/evidence.txt"
mq_protein_groups = "../../test_data/system_tests/organellar_maps/proteinGroups.txt"

directlfq_table = "../../test_data/system_tests/organellar_maps/evidence.txt.protein_intensities.tsv"

merged_table = "../../test_data/system_tests/organellar_maps/directLFQ_protein_intensities_merged.txt"
```

```python
import directlfq.lfq_manager as lfq_manager

lfq_manager.run_lfq(input_file=mq_evidence, input_type_to_use="maxquant_evidence_benchmark")

```

```python
#merge with relevant proteinGroups data
import pandas as pd

df_mq = pd.read_csv(mq_protein_groups, sep = "\t")
df_directlfq = pd.read_csv(directlfq_table, sep = "\t")



cols_to_drop = [x for x in df_mq.columns if "LFQ intensity" in x]
display(cols_to_drop)

df_mq_subset = df_mq.drop(columns = cols_to_drop)
df_mq_subset = df_mq_subset.astype({'id' : 'str'})
display(df_mq_subset)
display(df_directlfq)
```

```python
from matplotlib.pyplot import axis
import numpy as np

df_merged = df_directlfq.merge(right=df_mq_subset, left_on="protein", right_on='id', how='inner')
display(df_merged)
quant_columns = [x for x in df_directlfq.columns if "Map" in x and not "MS/MS" in x]
df_merged[quant_columns] = df_merged[quant_columns].replace({np.nan, 0})
print(quant_columns)

for col in quant_columns:
    df_merged[col] = df_merged[col].astype('int64')
rename_dict = {x : "LFQ intensity " + x for x in quant_columns}
df_merged = df_merged.rename(columns = rename_dict)
df_merged = df_merged.drop("protein", axis=1)

display(df_merged.columns)
df_merged.to_csv(merged_table, sep="\t", index=None)
```
